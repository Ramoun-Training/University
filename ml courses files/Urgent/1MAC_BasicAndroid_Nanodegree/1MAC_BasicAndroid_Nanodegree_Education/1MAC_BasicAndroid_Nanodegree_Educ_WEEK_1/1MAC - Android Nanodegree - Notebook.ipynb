 - **Syntax**:<br><br>
     Rules that define what a valid xml is.(if you write invalid xml, then it won't even show up on your phone).<br><br>
 - **xml syntax**:<br><br>         
     xml consists of elements -> an element consists of an openning tag,closing tag, and content.
     ```xml
         <ElementName attribute = "value">
            content
         </ElementName>
     ```
     if the element doesn't have a content we can use a shortcut
     ```xml
         <ElementName attribute = "value" />         
     ```
     <img src="C:\Users\Ramoun\Google Drive (omartarekramoun@gmail.com)\Urgent\1MAC_BasicAndroid_Nanodegree\1MAC_BasicAndroid_Nanodegree_Images\xml.png"/>
 
 - **how to write xml code**:<br><br>         
     - we always start with an openning angle bracket followed by a viewname(image view, button...etc), then we have a list of attributes one per line, then we have a forward slash and a closing angle bracket.
     - in xml, if you open a tag you always need to close it.
 - **self-closing tags**:<br><br>     
     - tags that doesn't need a closing tag(instead we use a forward slash in front of the right angle bracket of the openning tag).
     - example of a normal tag(seprate openning tag and seprate closing tag)
         <img src="C:\Users\Ramoun\Google Drive (omartarekramoun@gmail.com)\Urgent\1MAC_BasicAndroid_Nanodegree\1MAC_BasicAndroid_Nanodegree_Images\normal tag.png"/>
     - example of a self closing tag
         <img src="C:\Users\Ramoun\Google Drive (omartarekramoun@gmail.com)\Urgent\1MAC_BasicAndroid_Nanodegree\1MAC_BasicAndroid_Nanodegree_Images\xml.png"/>
     - __**we use self closing tags if the tag doesn't have content**__
 - **nesting elements**:<br><br>
     nesting is possible in xml(like in html)
     - example of nesting in xml
         <img src="C:\Users\Ramoun\Google Drive (omartarekramoun@gmail.com)\Urgent\1MAC_BasicAndroid_Nanodegree\1MAC_BasicAndroid_Nanodegree_Images\nesting.png"/>
     - closing tags help us organise the hierarchy of the xml document(who is the parent and who is the child).
